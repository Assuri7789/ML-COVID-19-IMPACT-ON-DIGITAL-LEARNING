```python
missing_val_count_by_column = (data.isnull().sum())
print(missing_val_count_by_column[missing_val_count_by_column > 0])
```

```python
removednull = data.fillna(data.mean())
```

```python
X = removednull.drop(['Product_Name'], axis=1)

y = removednull['Product_Name']
```

```python

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.33, random_state = 0)
```

```python
from sklearn.tree import DecisionTreeClassifier

model = DecisionTreeClassifier()
model.fit(X_train, y_train)
```

```python
def future_products(time,lp_id,pct_access,engagement_index,id,state,locale,pct_black,pct_free,couty_connections_ratio,pp_total_raw,URL,Provider,Sectors,Category,Subcategory):
    x = np.array([time,lp_id,pct_access,engagement_index,id,state,locale,pct_black,pct_free,couty_connections_ratio,pp_total_raw,URL,Provider,Sectors,Category,Subcategory])
    prediction = model.predict(x.reshape(1, -1))

    return prediction
```

```python
outputs = gr.Textbox()

app = gr.Interface(fn=future_products, inputs=['number','number','number','number','number','number','number','number','number','number','number','number','number','number','number','number'], outputs= "number",description="This is a diabetes model")
```

```python
app.launch()
#To provide a shareable link
app.launch(share = True)
```
